```python
duplicates = df[df.duplicated()]
print(len(duplicates))
duplicates.head()
```

```python
df["basement"] = df["sqft_basement"].apply(lambda x: False if x == "0.0" else True)
df.head()
```

```python
df["sqft_basement"].replace(to_replace="?", value="0.0", inplace=True)
df["sqft_basement"].unique()
```

```python
df["sqft_basement"].head()
```

```python
df["sqft_basement"] = df["sqft_basement"].astype(float, inplace=True)
df["sqft_basement"].mean()
```

```python
type(df["sqft_basement"])
```

```python
df.info()
```

```python
df.describe()
```

```python
df["sqft_living15"].isna().value_counts()
```

```python
df["yr_renovated"].fillna(value = 0)
df["yr_renovated"].isna()
```

```python
df["sqft_living15"].dtype
```

```python
df['was_renovated'] = df['sqft_living'].map(lambda x: "Yes" if x > df["sqft_living15"] else "No")
df.head()
```

```python
df["waterfront"].unique()
```

```python
df["waterfront"].value_counts()
```

```python
df["waterfront_view"] = df["waterfront"].map(lambda x: True if x > 0 else False)
```

```python
df["waterfront_view"].value_counts()
```

```python
df.drop["waterfront"]
df.head()
```

```python
df = df.drop(columns =  "waterfront")
df.head()
```

```python
df["bedrooms"].value_counts()
```

```python
df["renovated"] = df["sqft_living"] - df["sqft_living15"]
df.head()
```

```python
df["renovated"] = df["renovated"].map(lambda x: False if x == 0 else True)
```

```python
df["renovated"].value_counts()
```

```python
df = df.drop(columns = "yr_renovated")
df.head()
```

```python
df["percent_lot_used"] = (df["sqft_lot"] - df["sqft_living"]) / df["sqft_lot"]
df.head()
```

```python
df = df.drop(columns = ["lat", "long"])
df.head()
```

```python
df["age_of_house"] = 2019 - df["yr_built"]
df.head()
```

```python
df["upstairs_as_percent_of_house"] = (df["sqft_living"] - df["sqft_basement"]) / df["sqft_living"]
```
